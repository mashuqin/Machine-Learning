### 使用sklearn构造树模型

```python
#使用sklearn.tree可做分类、回归（即预测类别值、连续值均可以）
from sklearn import tree
#实例化树模型，传入参数(常用的有max_depth、min_samples_split)
#print(help(tree.DecisionTreeRegressor))
dtr=tree.DecisionTreeRegression(max_depth=2)
#
dtr.fit(housing.data[:,[6,7]],housing.target)
```
